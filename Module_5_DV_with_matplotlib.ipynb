# **Matplotlib Architecture**
- Provides functions and tools for representing and manipulating a Figure (main object) and its internal components.
- Includes event handling and animation capabilities for interactive charts (e.g., triggered by keyboard or mouse events).
- The architecture of matplotlib is logically structured in three layers placed at three different levels
- The communication is unidirectional, that is, each layer is able to communicate with the underlying layer, while the lower cannot do it with the top one.
- The three layers are as follows:
    - scripting
    - artist
    - backend

**Backend Layer**

- In the diagram of the matplotlib architecture the layer that works at the lowest level is the Backend.
- In this layer there are the matplotlib APIs, a set of classes that play the role of implementation of the graphic elements at a low level.
    - FigureCanvas is the object that embodies the concept of drawing area.
    - Renderer is the object that draws on FigureCanvas.
    - Event is the object that handles user inputs (keyboard and mouse events).

**Artist Layer**
- As an intermediate layer we have a layer called Artist.
- All that you can look inside a figure is an Artist object,
that is, all the elements that go to make up a chart such as the title, axis labels, markers, etc., are instances of
the Artist object.
- Each of these instances within the chart plays its role within a hierarchical structure

Screen Shot 2024-12-17 at 09.14.02.png

- There are two Artist classes: **primitive** and **composite**.
- The primitive artists are individual objects that constitute the basic elements to form a graphical
representation in a plot.
- For example a Line2D, or as a geometric figure such as a Rectangle or Circle, or even
pieces of text.
- The composite artists are those graphic elements present in a chart which are composed of several
base elements, namely, the primitive artists.
- Composite artists are for example the Axis, Ticks, Axes, and
figures.

- The three main Artist objects (composite
artists) that are generally used in all implementations:
    - Figure:
      -  Top-level object in the hierarchy.
      - Represents the entire graphical output.
      - Can contain multiple Axes objects.
    - Axes
      - Represents a plot or chart within a Figure.
      - Each Axes belongs to a single Figure.
      - Includes Artist Axis objects (two for 2D, three for 3D).
      - Other components, such as the title, x-label, and y-label, are part of this object.
    - Axis
      - Handles numerical values displayed on Axes.
      - Defines axis limits, tick positions, and tick labels.
      - Tick positions are managed by the Locator object.
      - Tick label formatting is handled by the Formatter object.

**Scripting Layer**
- Best suited for data analysis and visualization tasks.
- Consists of the pyplot module, which provides an easy-to-use scripting interface.

**Plotting Multiple Trends:**

- You can plot multiple trends (such as a sinusoidal function) on the same plot.
- Use NumPy's arange() to create a series of points on the x-axis.
- Use Python's sin() function from the math module to generate corresponding y-axis values.

```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Create the time array
t = np.arange(0, 2.5, 0.1)

# Use map and convert to list
y1 = list(map(lambda x: math.sin(math.pi * x), t))
y2 = list(map(lambda x: math.sin(math.pi * x + math.pi/2), t))
y3 = list(map(lambda x: math.sin(math.pi * x - math.pi/2), t))

# Plot the data
plt.plot(t, y1, 'b*', label='sin(t)')
plt.plot(t, y2, 'g^', label='sin(t + /2)')
plt.plot(t, y3, 'ys', label='sin(t - /2)')

# Add a title and labels
plt.title('Multiple Sinusoidal Functions')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')

# Show the legend
plt.legend()

# Display the plot
plt.show()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

# Generate x-axis data
t = np.arange(0, 2.5, 0.1)  # Array of values from 0 to 2.5 with a step of 0.1

# Generate y-axis data
y1 = list(map(lambda x: math.sin(math.pi * x), t))  # Sinusoidal trend
y2 = list(map(lambda x: math.sin(math.pi * x + math.pi / 2), t))  # Phase-shifted by 90
y3 = list(map(lambda x: math.sin(math.pi * x - math.pi / 2), t))  # Phase-shifted by -90

# Plot all three trends
plt.plot(t, y1, 'b--', label='y1: blue dashed')
plt.plot(t, y2, 'g', label='y2: green solid')
plt.plot(t, y3, 'r-.', label='y3: red dash-dot')

# Add labels, legend, and show plot
plt.xlabel('Time (t)')
plt.ylabel('Values')
plt.title('Sinusoidal Trends with Different Line Styles')
plt.legend()
plt.show()
```

# Using the kwargs

**Attributes and Customization:**

- The objects in a Matplotlib chart (like lines, axes, markers) have attributes that define their properties (e.g., color, line style, marker type).
- These attributes have default values but can be customized using kwargs (keyword arguments).

What Are kwargs?:

- kwargs are a flexible way to pass additional named arguments to functions.
- In Matplotlib, they are used to specify attributes of the plot elements, such as color (color), line width (linewidth), and markers (marker).

Function Signature:

The plot() function is represented as:

    matplotlib.pyplot.plot(*args, **kwargs)


```python
import matplotlib.pyplot as plt

# Data for the plot
y = [1, 2, 4, 2, 1, 0, 1, 2, 1, 4]

# Create a plot with customized line width
plt.plot(y, linewidth=10.0)  # Set line thickness to 2.0

# Add title and labels for clarity
plt.title("Line Plot with Custom Line Width")
plt.xlabel("Index")
plt.ylabel("Value")

# Display the plot
plt.show()
```

**Working with Multiple Figures and Axes**
Multiple Figures:
- You can create and manage multiple figures simultaneously in Matplotlib.
- Subplots: Subdivide a figure into multiple sections using the subplot() function, each acting as an independent plot (axes).
- Current Figure and Axes: Commands affect the current figure and current axes. Subplots focus commands on a specific area.
- subplot() Arguments:
  - The function takes three integers:
    1. Rows: Number of vertical divisions in the figure.
    2. Columns: Number of horizontal divisions.
    3. Index: Selects the current subplot to direct commands.


```python
import numpy as np
import matplotlib.pyplot as plt

# Data
t = np.arange(0, 5, 0.1)
y1 = np.sin(2 * np.pi * t)
y2 = np.sin(2 * np.pi * t)

# Create subplots
plt.subplot(211)  # First subplot (2 rows, 1 column, 1st plot)
plt.plot(t, y1, 'b-.')  # Blue dash-dot line
plt.title("Sine Wave - Top")

plt.subplot(212)  # Second subplot (2 rows, 1 column, 2nd plot)
plt.plot(t, y2, 'r--')  # Red dashed line
plt.title("Sine Wave - Bottom")

# Adjust layout for better display
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
t = np.arange(0., 1., 0.05)  # Time points
y1 = np.sin(2 * np.pi * t)  # Sine wave
y2 = np.cos(2 * np.pi * t)  # Cosine wave

# First subplot (Sine wave)
plt.subplot(121)  # 1 row, 2 columns, first plot
plt.plot(t, y1, 'b-.')  # Blue dashed-dot line for sine
plt.title("Sine Wave")  # Title for first subplot
plt.xlabel("Time (t)")  # X-axis label
plt.ylabel("Amplitude")  # Y-axis label
plt.grid(True)  # Add grid

# Second subplot (Cosine wave)
plt.subplot(122)  # 1 row, 2 columns, second plot
plt.plot(t, y2, 'r--')  # Red dashed line for cosine
plt.title("Cosine Wave")  # Title for second subplot
plt.xlabel("Time (t)")  # X-axis label
plt.ylabel("Amplitude")  # Y-axis label
plt.grid(True)  # Add grid

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plots
plt.show()
```

# **Adding Further Elements to the Chart**

**Adding Text**

```python
import matplotlib.pyplot as plt

# Set axis limits
plt.axis([0, 5, 0, 20])

# Add title and axis labels
plt.title('My first plot', fontsize=20)
plt.xlabel('Counting', color='gray')
plt.ylabel('Square values', color='gray')

# Add annotations for data points
plt.text(1, 1.5, 'First')
plt.text(2, 4.5, 'Second')
plt.text(3, 9.5, 'Third')
plt.text(4, 16.5, 'Fourth')

# Add an equation with styling
plt.text(1.1, 12, r'$y = x^2$', fontsize=20, bbox={'facecolor': 'yellow', 'alpha': 0.2})

# Plot data
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')

# Display the plot
plt.show()
```

# **Handling Date Values**
Handling date-time values in plots can be challenging, especially when dealing with large datasets or when dates are used along the x-axis. The primary issue is that dates are not inherently spaced evenly like numerical values, and displaying them as ticks can result in overlapping or unreadable labels.

```python
import datetime
import numpy as np
import matplotlib.pyplot as plt

# Define the dates for the x-axis (events)
events = [
    datetime.date(2015, 1, 23), datetime.date(2015, 1, 28),
    datetime.date(2015, 2, 3), datetime.date(2015, 2, 21),
    datetime.date(2015, 3, 15), datetime.date(2015, 3, 24),
    datetime.date(2015, 4, 8), datetime.date(2015, 4, 24)
]

# Define the readings for the y-axis
readings = [12, 22, 25, 20, 18, 15, 17, 14]

# Plot the data
plt.plot(events, readings)

# Display the plot
plt.show()
```

1. Issue with Automatic Date Ticks:
    - The default behavior of matplotlib for handling date ticks is often problematic.
    - The automatic tick labels on the x-axis can be hard to read and can overlap.
    - The time intervals between data points are not clearly represented, leading to a cluttered and confusing chart.
2. Using matplotlib.dates for Date Handling:
    - To handle date data effectively, you need to import the matplotlib.dates module, which is specialized for managing date and time values on plots.
3. Defining Time Scales:
    - You can define the scales for dates using MonthLocator() for months and DayLocator() for days. This allows you to control how dates are spaced and displayed on the x-axis.
4. Formatting Tick Labels:
    - Proper formatting is crucial to avoid overlap and ensure clarity.
    - Use the DateFormatter() function to define the format of the tick labels (e.g., "year-month").
    - Limiting the format to just the essential information (e.g., year and month) helps keep the chart clean.
5. Setting Major and Minor Ticks:
    - Once the scales are defined, you can control the appearance of ticks using set_major_locator() for major ticks (e.g., months) and set_minor_locator() for minor ticks (e.g., days).
    - These functions allow you to customize how dates appear on the x-axis.
6. Setting Tick Label Formats:
    - The set_major_formatter() function is used to specify the format of the major tick labels (e.g., displaying year-month format).


```python
import datetime
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Define the date format and tick locators
months = mdates.MonthLocator()  # Major ticks: months
days = mdates.DayLocator()      # Minor ticks: days
timeFmt = mdates.DateFormatter('%Y-%m')  # Format for tick labels

# Data points (dates and readings)
events = [
    datetime.date(2015, 1, 23), datetime.date(2015, 1, 28),
    datetime.date(2015, 2, 3), datetime.date(2015, 2, 21),
    datetime.date(2015, 3, 15), datetime.date(2015, 3, 24),
    datetime.date(2015, 4, 8), datetime.date(2015, 4, 24)
]
readings = [12, 22, 25, 20, 18, 15, 17, 14]

# Create the plot
fig, ax = plt.subplots()
plt.plot(events, readings)

# Set major and minor ticks, and format the x-axis
ax.xaxis.set_major_locator(months)  # Set major ticks to months
ax.xaxis.set_major_formatter(timeFmt)  # Set major tick labels format
ax.xaxis.set_minor_locator(days)  # Set minor ticks to days

# Display the plot
plt.show()
```
